```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=['Your Academic Stage','Do you have any bad habits like smoking, drinking on a daily basis?','Study Environment']
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['What coping strategy you use as a student?'],dtype=int)
df
```

```python
for col in df.columns:
    print(repr(col))
```

```python
final = df[['Your Academic Stage',
            'Peer pressure',
            'Academic pressure from your home',
            'Study Environment',
            'Do you have any bad habits like smoking, drinking on a daily basis?',
            'What would you rate the academic  competition in your student life',
            'Month',
            'hour',
            'What coping strategy you use as a student?_Analyze the situation and handle it with intellect',
            'What coping strategy you use as a student?_Emotional breakdown (crying a lot)',
            'What coping strategy you use as a student?_Social support (friends, family)',
            'Rate your academic stress index ']]
final
```

```python
corr = final.corr(numeric_only=True)
print(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
